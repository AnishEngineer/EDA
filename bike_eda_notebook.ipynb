# EDA - 1: Bike Dataset

This notebook contains solutions to the assignment questions using Python.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv('bike.csv')
df.head()
```

## Q1. Range of Selling Price

```python
min_price = df['selling_price'].min()
max_price = df['selling_price'].max()
print('Range:', max_price - min_price)
```

## Q2. Median Selling Price

```python
df['selling_price'].median()
```

## Q3. Most Common Seller Type

```python
df['seller_type'].value_counts()
```

## Q4. Bikes Driven > 50,000 km

```python
len(df[df['km_driven'] > 50000])
```

## Q5. Average km_driven per Owner Type

```python
df.groupby('owner')['km_driven'].mean()
```

## Q6. Proportion of Bikes from 2015 or Older

```python
len(df[df['year'] <= 2015]) / len(df)
```

## Q7. Missing Values

```python
df.isnull().sum()
```

## Q8. Highest Ex-showroom Price

```python
df.loc[df['ex_showroom_price'].idxmax()]
```

## Q9. Bikes per Seller Type

```python
df['seller_type'].value_counts()
```

## Q10. Price vs KM Driven (1st Owner)

```python
first_owner = df[df['owner'] == '1st owner']
sns.scatterplot(x='km_driven', y='selling_price', data=first_owner)
plt.show()
```
